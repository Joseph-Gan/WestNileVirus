```python
class ConvertDate(BaseEstimator,TransformerMixin):
    
    def fit(self, X, y=None):
        return self
    
    @property
    def getFeature(self):
        return self.feature
    
    def transform(self, X, y=None):
        X['date'] = pd.to_datetime(X.index)
        X['year'] = X.date.dt.year
        X['month'] = X.date.dt.month
        X['weekday'] = X.date.dt.weekday
        X = X.drop('date',axis=1)
        self.feature = X.columns.to_list()
        
        return X
    
class Calculate_Distance(BaseEstimator, TransformerMixin):
    
    """
    This class returns the distance between centroids and input latitude & longtitude 
        
    """
  
    def __init__(self, column_lat, column_lon, *, centroid_lat, centroid_lon, return_col_as):
        self.centroid_lat = centroid_lat
        self.centroid_lon = centroid_lon
        self.column_lat = column_lat
        self.column_lon = column_lon
        self.return_col_as = return_col_as

    @property
    def getFeature(self):
        return self.feature
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        distances = []
        
        for lat, lon in zip(X[self.column_lat], X[self.column_lon]):
            dis = distance((lat, lon), (self.centroid_lat, self.centroid_lon)).km
            distances.append(dis)
            
        X[f'{self.return_col_as}_km'] = distances
        
        self.feature = X.columns.to_list()
        
        return X

class RareLabelEncoding(BaseEstimator, TransformerMixin):
    
    
    """
    This class maps the string value to numerical as per input dictionary.
    
    This this project. We map as following.
    {'CULEX PIPIENS':3, 
    'CULEX PIPIENS/RESTUANS':2, 
    'CULEX RESTUANS': 1},
    'RARE' = 0), 
    
    """
    
    
    def __init__(self, variable, *, map_to, replace_with):
        self.variable = variable
        self.replace_with = replace_with
        self.map_to = map_to
        
    @property
    def getFeature(self):
        return self.feature
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        
        X[f'{self.variable}_tr'] = np.where(X[self.variable].isin(self.map_to),
                                            X[self.variable].map(self.map_to),
                                            self.replace_with)
        X = X.drop([self.variable], axis=1)
        
        self.feature = X.columns.to_list()
        
        return X
```

# Feature Engineering

```python
#ranking of mosquito species according to the finding of EDA
mapping = {'CULEX PIPIENS':3, 'CULEX PIPIENS/RESTUANS':2, 'CULEX RESTUANS': 1}

#general pipeline for feature engineering 
transformer_pipe = make_pipeline(
    
    #convert date to year,month,day
    ConvertDate(),
    
    #calculate distance between centroid_1
    Calculate_Distance('latitude', 'longitude', 
                       centroid_lat=41.95726979, centroid_lon=-87.78117394,
                       return_col_as = 'dist_centroid_1'),
    
    #calculate distance between centroid_2
    Calculate_Distance('latitude', 'longitude', 
                       centroid_lat=41.74808405, centroid_lon=-87.65556425,
                       return_col_as = 'dist_centroid_2'),
    
    #encode species as per mapping
    RareLabelEncoding('species', 
                        map_to = mapping,
                        replace_with = 0)
    )

pipe_tr = make_union(transformer_pipe)
```

# With Cost Sensitive Learning

There are 2 ways in which we can introduce cost into the learning function of the algorithm with Scikit-learn:

Defining the **class_weight** parameter for those estimators that allow it, when we set the estimator.
For those who dont support **class_weight**, passing a **sample_weight** vector with the weights for every single observation, when we fit the estimator.

```python
def run_model2(model, name, X, y, balance):
    
    #create pipeline
    pipe = imb_make_pipeline(pipe_tr, model)
    
    #setting sampling weight for algorithm that does not support class_weight
    if name in ['adaboostclassifier', 'gradientboostingclassifier']:
      
        param = {f'{name}__sample_weight' : balance}
        
    else:
        param = None
        
        
    return cross_val_score(pipe, X, y, cv=5, scoring = 'roc_auc', fit_params=param).mean()
```

# Feature Selection

We will use **Recursive feature Elimination using Embedded Methods (Random Forest Feature Importance)**

This method consists of the following steps:

1) Rank the features according to their importance derived from a machine learning algorithm: it can be tree importance or coefficients derived from linear models.

2) Remove one feature -the least important- and build a machine learning algorithm utilising the remaining features.

3) Calculate a performance metric of choice: roc-auc, mse, rmse, accuracy, etc. 

4) If the metric decreases by more of an arbitrarily set threshold, then that feature is important and should be kept. Otherwise, we can remove that feature.

5) Repeat steps 2-4 until all features have been evaluated

We will use SKlearn RFECV that does above steps

## Baseline:Training Selected ML Models by fitting all feature

```python
#function to rank feature importance

def feature_importance(name, model, *, rfecv=False):
    
    #getting the feature importance
    features = pd.Series(model[name].feature_importances_)
    
    #getting the index of feature
    col = model['featureunion'].transformer_list[0][1].named_steps.rarelabelencoding.getFeature
    
    
    if rfecv:
        
        support_ = model['rfecv'].support_
        
        features.index = [col for col , boolean in zip(col, support_) if boolean]
    else:
        features.index = col

    # sort the features by importance
    features.sort_values(ascending=False, inplace=True)

#     # plot
#     ax = features.plot.bar(figsize=(20,6))
#     ax.set_title('name')
#     ax.set_ylabel('Importance')
#     ax.set_xlabel('Feature')
    return features
```

ROC_AUC is a metric to compare performance between models to separate the True Positive and False Positive class that operates in all threshold.

For business case, we generally operate on a single threshold to derive the recall/precision. Hence we shall find out the best threshold to operate and plot the confusion matrix.

```python
def plotlearningcurve(estimator, title, X, y, ylim=None, cv=5, n_jobs=None, train_sizes = np.linspace(.1, 1, 10), scoring=None):
    
    _, axes = plt.subplots(1, 1, figsize=(20, 5))
    axes.set_title(title)
    axes.set_xlabel('Training Examples')
    axes.set_ylabel('Scores')
    
    train_sizes_abs, train_scores, cv_scores = learning_curve(estimator, X, y, cv=cv, n_jobs= n_jobs, scoring=scoring, train_sizes=train_sizes)
    
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    
    cv_scores_mean = np.mean(cv_scores, axis=1)
    cv_scores_std = np.std(cv_scores, axis=1)
    
    axes.grid()
    
    axes.plot(train_sizes_abs, train_scores_mean, 'o-', color='r', label="Training score")
    axes.plot(train_sizes_abs, cv_scores_mean, 'o-', color='b', label="CV score")
    
    axes.fill_between(train_sizes_abs, 
                     train_scores_mean + train_scores_std, 
                     train_scores_mean - train_scores_std,
                    alpha = 0.1,
                    color='r')

    axes.fill_between(train_sizes_abs, 
                 cv_scores_mean + cv_scores_std, 
                 cv_scores_mean - cv_scores_std,
                alpha = 0.1,
                color='b')
    

    
    axes.legend(loc='best')
    
    return plt
```

## Best threshold

```python
#generating y_score with validation set.
y_score = final_model.predict_proba(X_validation_tr)[:, 1]

#returning sets of precision, and recall for a given sets of threshold.
precision, recall, threshold = precision_recall_curve(y_validation, y_score);
```

```python
#calculating fbeta score
def fbeta_score(precision, recall, beta):
    nominator = (1 + beta**2) * (precision * recall)
    denominator = (beta**2 * precision) + recall
    
    return nominator / denominator
```

```python
#calculating fbeta score for all threshold
fbeta =  fbeta_score(precision, recall, 1)
fbeta_training = np.max(fbeta)
#returning the threshold of maximum fbeta
best_threshold = threshold[np.argmax(fbeta)]
```

```python
print(f"best threshold: {best_threshold}")
print(f"f1 score = {fbeta_training}")
```

```python
#plotting confusion matrix

#convert probability to true label
y_pred = pd.Series([1 if prob >= best_threshold else 0 for prob in y_score])


cm = confusion_matrix(y_validation, y_pred)

ConfusionMatrixDisplay(confusion_matrix=cm,
                            display_labels=['WNV Not Present', 'WNV Present']).plot(cmap='GnBu');
```
