```python
class ConvertDate(BaseEstimator,TransformerMixin):
    
    def fit(self, X, y=None):
        return self
    
    @property
    def getFeature(self):
        return self.feature
    
    def transform(self, X, y=None):
        X['date'] = pd.to_datetime(X.index)
        X['year'] = X.date.dt.year
        X['month'] = X.date.dt.month
        X['weekday'] = X.date.dt.weekday
        X = X.drop('date',axis=1)
        self.feature = X.columns.to_list()
        
        return X
    
class Calculate_Distance(BaseEstimator, TransformerMixin):
    
    """
    This class returns the distance between centroids and input latitude & longtitude 
        
    """
  
    def __init__(self, column_lat, column_lon, *, centroid_lat, centroid_lon, return_col_as):
        self.centroid_lat = centroid_lat
        self.centroid_lon = centroid_lon
        self.column_lat = column_lat
        self.column_lon = column_lon
        self.return_col_as = return_col_as

    @property
    def getFeature(self):
        return self.feature
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        distances = []
        
        for lat, lon in zip(X[self.column_lat], X[self.column_lon]):
            dis = distance((lat, lon), (self.centroid_lat, self.centroid_lon)).km
            distances.append(dis)
            
        X[f'{self.return_col_as}_km'] = distances
        
        self.feature = X.columns.to_list()
        
        return X

class RareLabelEncoding(BaseEstimator, TransformerMixin):
    
    
    """
    This class maps the string value to numerical as per input dictionary.
    
    This this project. We map as following.
    {'CULEX PIPIENS':3, 
    'CULEX PIPIENS/RESTUANS':2, 
    'CULEX RESTUANS': 1},
    'RARE' = 0), 
    
    """
    
    
    def __init__(self, variable, *, map_to, replace_with):
        self.variable = variable
        self.replace_with = replace_with
        self.map_to = map_to
        
    @property
    def getFeature(self):
        return self.feature
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        
        X[f'{self.variable}_tr'] = np.where(X[self.variable].isin(self.map_to),
                                            X[self.variable].map(self.map_to),
                                            self.replace_with)
        X = X.drop([self.variable], axis=1)
        
        self.feature = X.columns.to_list()
        
        return X
```

# Import DataFrame 

```python
df = pd.read_csv("./asset/full_train.csv", index_col ='Unnamed: 0')
```

# Train test Split

It is important to train test split before any feature engineering to avoid data leakage. We will use kaggle test.csv as our test set for final model evaluation. Hence here, we will split into train and validation instead.

```python
X = df.drop(['trap', 'nummosquitos', 'wnvpresent'], axis=1)
y = df['wnvpresent']

X_train, X_validation, y_train, y_validation = train_test_split(X, y, stratify=y, test_size=0.33, random_state=42)
```

# Feature Engineering

```python
#ranking of mosquito species according to the finding of EDA
mapping = {'CULEX PIPIENS':3, 'CULEX PIPIENS/RESTUANS':2, 'CULEX RESTUANS': 1}

#general pipeline for feature engineering 
transformer_pipe = make_pipeline(
    
    #convert date to year,month,day
    ConvertDate(),
    
    #calculate distance between centroid_1
    Calculate_Distance('latitude', 'longitude', 
                       centroid_lat=41.95726979, centroid_lon=-87.78117394,
                       return_col_as = 'dist_centroid_1'),
    
    #calculate distance between centroid_2
    Calculate_Distance('latitude', 'longitude', 
                       centroid_lat=41.74808405, centroid_lon=-87.65556425,
                       return_col_as = 'dist_centroid_2'),
    
    #encode species as per mapping
    RareLabelEncoding('species', 
                        map_to = mapping,
                        replace_with = 0)
    )

pipe_tr = make_union(transformer_pipe)
```

# Model

As a baseline, we will train model with all features & with default hyperparameter setting.

```python
# setting global random seed for reproducibility
rng = np.random.RandomState(0)
```

```python
def run_model(balancing, model, X, y):
    pipe = imb_make_pipeline(pipe_tr, balancing, model)
    return cross_val_score(pipe, X, y, cv=5, scoring = 'roc_auc').mean()
```

```python
models_dict ={
    'rf' : RandomForestClassifier(random_state=rng, n_jobs=-1),
    'xgb' : XGBClassifier(random_state=rng, n_jobs=-1, use_label_encoder=False, eval_metric = 'auc'),
    'adaboost' : AdaBoostClassifier(random_state=rng),
    'gradientboost' : GradientBoostingClassifier(random_state=rng)
}
```

## With Over Sampler

```python
oversampler_dict = {

    'random_oversampler': RandomOverSampler(
        sampling_strategy='auto',
        random_state=rng),

    'smote': SMOTE(
        sampling_strategy='auto',  # samples only the minority class
        random_state=rng,  # for reproducibility
        k_neighbors=5,
        n_jobs=4),

}
```

```python
total_result = {}
for model_name, model in models_dict.items():
    for sampler_name, sampler in oversampler_dict.items():

        result = run_model(sampler, model, X_train, y_train)

        total_result[f'{model_name}_{sampler_name}'] = result
        print(f'{model_name}_{sampler_name}: {result:.2f}')
```

# With Cost Sensitive Learning

There are 2 ways in which we can introduce cost into the learning function of the algorithm with Scikit-learn:

Defining the **class_weight** parameter for those estimators that allow it, when we set the estimator.
For those who dont support **class_weight**, passing a **sample_weight** vector with the weights for every single observation, when we fit the estimator.

```python
def run_model2(model, name, X, y, balance):
    
    #create pipeline
    pipe = imb_make_pipeline(pipe_tr, model)
    
    #setting sampling weight for algorithm that does not support class_weight
    if name in ['adaboostclassifier', 'gradientboostingclassifier']:
      
        param = {f'{name}__sample_weight' : balance}
        
    else:
        param = None
        
        
    return cross_val_score(pipe, X, y, cv=5, scoring = 'roc_auc', fit_params=param).mean()
```

```python
models_dict2 ={
    'rf' : RandomForestClassifier(random_state=rng, n_jobs=-1, class_weight ='balanced'),
    'xgb' : XGBClassifier(random_state=rng, n_jobs=-1, use_label_encoder=False, eval_metric = 'auc', scale_pos_weight=17),
    'adaboostclassifier' : AdaBoostClassifier(random_state=rng),
    'gradientboostingclassifier' : GradientBoostingClassifier(random_state=rng),
}
```

```python
balance = np.where(y_train==1,17,1)

total_result2 = {}
for model_name, model in models_dict2.items():
    
        
        result = run_model2(model, model_name, X_train, y_train, balance)
        
        
        total_result2[f'{model_name}_cost_sensitive_learning'] = result
        print(f'{model_name}_cost_sensitive_learning : {result:.2f}')


```

```python
#plotting the result in plot
result_table = pd.DataFrame.from_dict({**total_result, **total_result2}, orient='index', columns = ['roc_auc'])
plt.figure(figsize=(10,5))
ax = sns.scatterplot(x=result_table.index, y=result_table.roc_auc, color='red', s=50)
plt.xticks(rotation=90)
ax.set_ylim([0.6, 1]);
```

**Obervation:**

For the balancing method, there is no significant difference between **RandomOverSampling, Smote or Cost Learning Function**
We will choose cost learning function as the balancing method.

We see that the result is close and it is hard to tell which algorithm gives better results after hyper-parameter tuning. However, scalablity of xgboost is be faster than gradientboost as xgboost uses second order derivative to calculate loss function while gradientboost uses first order derivative. Hence, we will choose xgboost and randomforest for further feature selection & hyperparameter tuning.


# Feature Selection

We will use **Recursive feature Elimination using Embedded Methods (Random Forest Feature Importance)**

This method consists of the following steps:

1) Rank the features according to their importance derived from a machine learning algorithm: it can be tree importance or coefficients derived from linear models.

2) Remove one feature -the least important- and build a machine learning algorithm utilising the remaining features.

3) Calculate a performance metric of choice: roc-auc, mse, rmse, accuracy, etc. 

4) If the metric decreases by more of an arbitrarily set threshold, then that feature is important and should be kept. Otherwise, we can remove that feature.

5) Repeat steps 2-4 until all features have been evaluated

We will use SKlearn RFECV that does above steps

## Baseline:Training Selected ML Models by fitting all feature

```python
#function to rank feature importance

def feature_importance(name, model, *, rfecv=False):
    
    #getting the feature importance
    features = pd.Series(model[name].feature_importances_)
    
    #getting the index of feature
    col = model['featureunion'].transformer_list[0][1].named_steps.rarelabelencoding.getFeature
    
    
    if rfecv:
        
        support_ = model['rfecv'].support_
        
        features.index = [col for col , boolean in zip(col, support_) if boolean]
    else:
        features.index = col

    # sort the features by importance
    features.sort_values(ascending=False, inplace=True)

#     # plot
#     ax = features.plot.bar(figsize=(20,6))
#     ax.set_title('name')
#     ax.set_ylabel('Importance')
#     ax.set_xlabel('Feature')
    return features
```

```python
#building base model pipeline
all_pipe = {
    'randomforestclassifier' : imb_make_pipeline(pipe_tr,
                                                 SMOTE( sampling_strategy='auto',  # samples only the minority class
                                                       random_state=rng,  # for reproducibility
                                                       k_neighbors=5,
                                                       n_jobs=4),
                                                 RandomForestClassifier(random_state=rng, n_jobs=-1, class_weight ='balanced')),
    
    'xgbclassifier' : imb_make_pipeline(pipe_tr,
                                         SMOTE( sampling_strategy='auto',  # samples only the minority class
                                               random_state=rng,  # for reproducibility
                                               k_neighbors=5,
                                               n_jobs=4),                                        
                                        XGBClassifier(random_state=rng, n_jobs=-1, use_label_encoder=False, eval_metric = 'auc')),
    
    
    'adaboostclassifier' : imb_make_pipeline(pipe_tr, 
                                         SMOTE( sampling_strategy='auto',  # samples only the minority class
                                               random_state=rng,  # for reproducibility
                                               k_neighbors=5,
                                               n_jobs=4),                                             
                                             AdaBoostClassifier(random_state=rng))}
```

```python
score = defaultdict(dict)

for name, model in all_pipe.items():
    

    model.fit(X_train, y_train)
    
    y_pred_train = model.predict_proba(X_train)[:, 1]
    y_pred_validation = model.predict_proba(X_validation)[:, 1]
    
    roc_score_train = roc_auc_score(y_train, y_pred_train)
    roc_score_test = roc_auc_score(y_validation, y_pred_validation)
    
    feature = feature_importance(name, model, rfecv=False)
    
    
    score[f'{name}_base']['Feature_Number'] = len(feature)
    score[f'{name}_base']['feature'] = feature
    score[f'{name}_base']['roc_score_train'] = roc_score_train
    score[f'{name}_base']['roc_score_test'] = roc_score_test
    
    
```

```python
summary = pd.DataFrame.from_dict(score, orient='index').drop('feature', axis=1)
summary
```

# Hyperparameter Tuning

At this section, we will tune XGBoost Classifier & RandomForest Classifier to deal with overfitting

## Hyperparameter Tuning

```python
xgb_params = {
    'subsample' : [0.5, 0.6, 0.7],
    'learning_rate' : [0.0001, 0.001, 0.01, 0.1, 0.2, 0.3],
    'colsample_bytree' : [0.3, 0.5, 0.7, 1.0],
    'min_child_weight' : [1,3,5,7],
    'max_depth' : [2, 3 , 4 , 5, 6],
    'n_estimators' : [100, 200, 300, 400, 500],
    'reg_alpha' : [500, 1000, 2000, 3000]
}

rf_params = {'max_depth' : [2, 3 , 4 , 5, 6],
         'min_samples_split' : [2, 3, 4, 5, 6, 7, 8, 9, 10],
         'bootstrap' : [True, False],
            
         }


ada_params = {'n_estimators' : [500, 100, 200, 500, 1000],
         }


param = {'randomforestclassifier' : rf_params,
         'xgbclassifier' : xgb_params,
         'adaboostclassifier' : ada_params
        
        }
```

```python
sel_feature = score['xgbclassifier_rfecv']['feature'].index
```

```python
smote =  SMOTE(sampling_strategy='auto',  # samples only the minority class
               random_state=rng,  # for reproducibility
               k_neighbors=5,
               n_jobs=4)
```

```python
X_train_tr = transformer_pipe.fit_transform(X_train)[sel_feature]
X_train_tr, y_train_tr = smote.fit_resample(X_train_tr, y_train)

X_validation_tr = transformer_pipe.transform(X_validation)[sel_feature]
```

```python
models = {
    'randomforestclassifier' : RandomForestClassifier(random_state=rng, n_jobs=-1),
    'xgbclassifier' : XGBClassifier(random_state=rng, n_jobs=-1, use_label_encoder=False, eval_metric = 'auc'),
    'adaboostclassifier' : AdaBoostClassifier(random_state=rng)
 
}
```

```python
best_models = {}
for name, model in models.items():
    
    grid = RandomizedSearchCV(model, param[name], scoring='roc_auc', verbose=1, n_iter=20)
    

    grid.fit(X_train_tr, y_train_tr)

    
    
    best_model = grid.best_estimator_
    best_models[f'{name}_tuned'] = best_model
    
    y_pred_train = best_model.predict_proba(X_train_tr)[:, 1]
    y_pred_validation = best_model.predict_proba(X_validation_tr)[:, 1]
    
    roc_score_train = roc_auc_score(y_train_tr, y_pred_train)
    roc_score_test = roc_auc_score(y_validation, y_pred_validation)
    
    
    score[f'{name}_tuned']['Feature_Number'] = len(sel_feature)
    score[f'{name}_tuned']['feature'] = sel_feature
    score[f'{name}_tuned']['roc_score_train'] = roc_score_train
    score[f'{name}_tuned']['roc_score_test'] = roc_score_test
    
```

```python
summary = pd.DataFrame.from_dict(score, orient='index').drop('feature', axis=1)
summary
```

# Final Model

```python
final_model = best_models['xgbclassifier_tuned']
```
